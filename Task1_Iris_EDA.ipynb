## 1. Import Libraries

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set global plot style for clean visuals
sns.set_theme(style="whitegrid", palette="Set2")
plt.rcParams["figure.figsize"] = (10, 6)
plt.rcParams["axes.titlesize"] = 14

print("All libraries imported successfully.")
```

## 2. Load the Dataset

```
# Load Iris dataset directly via seaborn (no manual download needed)
iris = sns.load_dataset("iris")

# Also available via: pd.read_csv("iris.csv")
print("Dataset loaded successfully.")
print(f"Shape: {iris.shape}")
print(f"\nColumn Names:\n{iris.columns.tolist()}")
```

## 3. Initial Inspection

```
# First few rows
print("=== First 5 Rows ===")
iris.head()
```

```
# Data types and non-null counts
print("=== Dataset Info ===")
iris.info()
```

```
# Descriptive statistics: count, mean, std, min, max, quartiles
print("=== Descriptive Statistics ===")
iris.describe().round(2)
```

```
# Check class distribution
print("=== Species Count ===")
print(iris['species'].value_counts())
```

## 4. Data Visualization

### 4.1 Scatter Plot – Feature Relationships

```
# Pairplot: shows all pairwise relationships colored by species
fig = sns.pairplot(iris, hue="species", diag_kind="kde", plot_kws={"alpha": 0.7})
fig.fig.suptitle("Pairplot of Iris Features by Species", y=1.02, fontsize=16)
plt.tight_layout()
plt.savefig("task1_pairplot.png", dpi=150, bbox_inches="tight")
plt.show()
print("Scatter/pairplot saved.")
```

```
# Focused scatter: Sepal Length vs Petal Length
plt.figure(figsize=(8, 5))
sns.scatterplot(data=iris, x="sepal_length", y="petal_length",
                hue="species", style="species", s=80, alpha=0.85)
plt.title("Sepal Length vs Petal Length by Species")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.legend(title="Species")
plt.tight_layout()
plt.savefig("task1_scatter.png", dpi=150, bbox_inches="tight")
plt.show()
```

### 4.2 Histograms – Feature Distributions

```
# Histogram for each numeric feature, split by species
features = ["sepal_length", "sepal_width", "petal_length", "petal_width"]
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.flatten()

for i, feature in enumerate(features):
    for species in iris["species"].unique():
        subset = iris[iris["species"] == species][feature]
        axes[i].hist(subset, bins=15, alpha=0.6, label=species)
    axes[i].set_title(f"Distribution of {feature.replace('_', ' ').title()}")
    axes[i].set_xlabel(feature.replace("_", " ").title() + " (cm)")
    axes[i].set_ylabel("Frequency")
    axes[i].legend(title="Species")

plt.suptitle("Feature Histograms by Species", fontsize=15)
plt.tight_layout()
plt.savefig("task1_histograms.png", dpi=150, bbox_inches="tight")
plt.show()
```

### 4.3 Box Plots – Outlier Detection

```
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.flatten()

for i, feature in enumerate(features):
    sns.boxplot(data=iris, x="species", y=feature, ax=axes[i],
                palette="Set2", width=0.5)
    axes[i].set_title(f"Box Plot: {feature.replace('_', ' ').title()}")
    axes[i].set_xlabel("Species")
    axes[i].set_ylabel(feature.replace("_", " ").title() + " (cm)")

plt.suptitle("Box Plots for Outlier Detection by Species", fontsize=15)
plt.tight_layout()
plt.savefig("task1_boxplots.png", dpi=150, bbox_inches="tight")
plt.show()
```

### 4.4 Correlation Heatmap

```
plt.figure(figsize=(7, 5))
corr = iris.drop(columns=["species"]).corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm",
            linewidths=0.5, square=True, cbar_kws={"shrink": 0.8})
plt.title("Feature Correlation Heatmap")
plt.tight_layout()
plt.savefig("task1_heatmap.png", dpi=150, bbox_inches="tight")
plt.show()
```
